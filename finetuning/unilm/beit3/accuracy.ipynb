```python
import csv

# ----------------------------------------------------------------------
# 설정 부분 (사용자 환경에 맞게 파일 이름을 수정해주세요)
# ----------------------------------------------------------------------

# 1. 예측 답안이 포함된 파일
PREDICTION_FILE = 'submission.csv' 

# 2. 실제 정답이 포함된 파일
GROUND_TRUTH_FILE = 'train.csv'

# ----------------------------------------------------------------------
# 코드 본문 (이 아래는 수정할 필요가 없습니다)
# ----------------------------------------------------------------------

def calculate_accuracy_by_order():
    """
    두 CSV 파일의 순서를 기준으로 answer를 비교하여 정확도를 계산합니다.
    """
    
    correct_count = 0
    total_count = 0
    
    try:
        # 두 파일을 동시에 엽니다.
        with open(PREDICTION_FILE, mode='r', encoding='utf-8') as pred_file, \
             open(GROUND_TRUTH_FILE, mode='r', encoding='utf-8') as truth_file:
            
            # 각 파일에 대해 CSV 리더를 생성합니다.
            pred_reader = csv.DictReader(pred_file)
            truth_reader = csv.DictReader(truth_file)
            
            print("두 파일의 데이터를 순서대로 비교합니다...")
            
            # zip을 사용하여 두 파일을 한 줄씩 동시에 읽습니다.
            # zip은 두 파일 중 더 짧은 쪽이 끝나면 자동으로 멈춥니다.
            for pred_row, truth_row in zip(pred_reader, truth_reader):
                total_count += 1
                
                # 'answer' 열의 값을 비교합니다.
                if pred_row['answer'] == truth_row['answer']:
                    correct_count += 1
                    
    except FileNotFoundError as e:
        print(f"오류: 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요. ({e.filename})")
        return
    except KeyError as e:
        print(f"오류: 파일에 'answer' 열이 없습니다. CSV 파일의 헤더(첫 줄)를 확인해주세요. (오류 발생 열: {e})")
        return
    except Exception as e:
        print(f"파일 처리 중 오류가 발생했습니다: {e}")
        return

    # --- 최종 정확도 계산 및 출력 ---
    print("\n--- 계산 결과 ---")
    if total_count > 0:
        accuracy = (correct_count / total_count) * 100
        print(f"전체 비교 데이터 수: {total_count}개")
        print(f"정답 개수: {correct_count}개")
        print(f"정확도(Accuracy): {accuracy:.2f}%")
    else:
        print("비교할 데이터가 없습니다. 파일이 비어있는지 확인해주세요.")

# 스크립트 실행
if __name__ == '__main__':
    calculate_accuracy_by_order()
```
