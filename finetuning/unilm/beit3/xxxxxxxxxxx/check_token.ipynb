```python
# Hugging Face의 transformers 라이브러리 설치 필요
# pip install transformers sentencepiece

from transformers import XLMRobertaTokenizer

# 'beit3.spm' 파일이 있는 경로를 정확하게 지정해야 합니다.
tokenizer_path = "/data/2_data_server/cv-07/dice/2025_samsung_challenge/unilm/beit3/beit3.spm" 
```

```python
tokenizer = XLMRobertaTokenizer(tokenizer_path)
```

```python
true_text = "true"
false_text = "false"
```

```python
true_token_id = tokenizer.encode(true_text)[1]
false_token_id = tokenizer.encode(false_text)[1]
```

```python
print(f"단어 'true'의 토큰 ID: {true_token_id}")
print(f"단어 'false'의 토큰 ID: {false_token_id}")
```

```python

# --- 2. 토큰화 과정을 단계별로 확인하기 (참고용) ---
print("\n--- 'true'의 토큰화 과정 ---")
true_tokens = tokenizer.tokenize(true_text)
print(f"1. 토큰으로 분리: {true_tokens}")

true_ids_from_tokens = tokenizer.convert_tokens_to_ids(true_tokens)
print(f"2. 토큰을 ID로 변환: {true_ids_from_tokens}")
```
