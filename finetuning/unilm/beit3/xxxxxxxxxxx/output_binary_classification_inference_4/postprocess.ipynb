```python
import pandas as pd
csv_path = "/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_binary_classification_inference_4/eval_results.csv"
```

```python
data = pd.read_csv(csv_path)
```

```python
logit = data["preds"].values
```

```python
submission = []
for i in range(0, len(logit),4):
    argmax = logit[i:i+4].argmax()
    if argmax == 0:
        answer = "A"
    elif argmax == 1:
        answer = "B"
    elif argmax == 2:
        answer = "C"
    elif argmax == 3:
        answer = "D"

    submission.append({
        "ID": f'TEST_{i//4:03d}',
        "answer": answer
    })
```

```python
submission
```

```python
# save submission
submission_path = "/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_binary_classification_inference_4/binary_classification_submission.csv"
submission_df = pd.DataFrame(submission)
submission_df.to_csv(submission_path, index=False)
print(f"Submission saved to {submission_path}")
```
