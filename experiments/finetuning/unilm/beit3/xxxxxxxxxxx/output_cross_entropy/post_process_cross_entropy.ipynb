```python
import json

json_path = "/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_cross_entropy/evaluation_logits.json"
with open(json_path, "r") as f:
    data = json.load(f)
```

```python
len(data)
```

```python
answers = []
for item in data:
    argmax_index = item.index(max(item))
    if argmax_index == 0:
        answers.append("A")
    elif argmax_index == 1:
        answers.append("B")
    elif argmax_index == 2:
        answers.append("C")
    elif argmax_index == 3:
        answers.append("D")
```

```python
# answers

# make submission file
submission = []
for i, answer in enumerate(answers):
    submission.append({
        "id": f"TEST{i:03d}",
          "answer": answer
    })
```

```python
submission
```

```python
csv_path = "/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_cross_entropy/submission.csv"
import pandas as pd
# save submission
pd.DataFrame(submission).to_csv(csv_path, index=False)
```
