```python
import pandas as pd

csv_path = "/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_language_model_tf_inference_4_lossweight/eval_results.csv"
```

```python
data = pd.read_csv(csv_path)
# data.head()
# data['true'] = data['true'][2:]
# data['false'] = data['false'][:-2]
data.head()
```

```python
# type(data[0])
```

```python
index = []

for i in range(0, len(data), 4):
    data_row = data['true'].iloc[i:i+4]
    print(data_row)
    max_index = data_row.idxmax() - 4*(i // 4)
    if max_index == 0:
        index.append('A')
    elif max_index == 1:
        index.append('B')
    elif max_index == 2:
        index.append('C')
    elif max_index == 3:
        index.append('D')
```

```python
index
```

```python
# make csv file
output_df = pd.DataFrame()
output_df['ID'] = [f'TEST_{i:03d}' for i in range(0, len(index))]
output_df['answer'] = index
```

```python
# save to csv
output_df.to_csv('/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_language_model_tf_inference_4_lossweight/language_modeling_tf.csv', index=False)
```
