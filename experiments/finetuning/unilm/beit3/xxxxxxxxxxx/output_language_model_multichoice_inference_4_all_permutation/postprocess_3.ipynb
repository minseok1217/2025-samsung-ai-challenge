```python
import pandas as pd

# 1. 입력 파일과 출력 파일 이름 설정
input_filename = '/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_language_model_multichoice_inference_4_all_permutation/eval_results.csv'  # 가지고 계신 csv 파일 이름으로 변경하세요.
output_filename = '/data/2_data_server/cv-07/dice/2025_samsung_challenge/finetuning/unilm/beit3/output_language_model_multichoice_inference_4_all_permutation/submission.csv'

try:
    # --- 📍 해결 코드 ---
    # pandas가 파일의 첫 줄을 헤더로 자동 인식하도록 옵션을 제거합니다.
    df = pd.read_csv(input_filename)
    # ---------------------

    # 데이터 타입을 숫자로 확실하게 지정 (데이터가 깨끗하다면 이 줄은 생략 가능)
    df = df.astype(float)

    # 각 행에서 최댓값을 가진 열의 이름을 찾기
    max_columns = df.idxmax(axis=1)

    # 제출 형식에 맞는 DataFrame 생성
    output_csv = pd.DataFrame()
    output_csv['ID'] = [f'TEST_{i:03}' for i in range(len(max_columns))]
    output_csv['answer'] = max_columns

    # 결과 확인 (상위 5개 행 출력)
    print("생성된 submission 파일 미리보기:")
    print(output_csv.head())

    # DataFrame을 CSV 파일로 저장
    output_csv.to_csv(output_filename, index=False)

    print(f"\n✅ '{output_filename}' 파일이 성공적으로 생성되었습니다.")

except FileNotFoundError:
    print(f"[오류] '{input_filename}' 파일을 찾을 수 없습니다. 파일 이름을 확인해주세요.")
except ValueError as e:
    print(f"[데이터 타입 변환 오류] CSV 파일에 숫자로 바꿀 수 없는 값이 있습니다: {e}")
except Exception as e:
    print(f"알 수 없는 오류가 발생했습니다: {e}")
```
