```python
import json
import csv

# ----------------------------------------------------------------------
# 설정 부분 (사용자 환경에 맞게 수정해주세요)
# ----------------------------------------------------------------------

# 1. 입력으로 사용할 JSON 파일 이름을 지정하세요. (점수들이 담긴 파일)
INPUT_JSON_FILE = 'score.json'

# 2. 결과가 저장될 CSV 파일 이름을 지정하세요.
OUTPUT_CSV_FILE = 'submission.csv'

# ----------------------------------------------------------------------
# 코드 본문 (이 아래는 수정할 필요가 없습니다)
# ----------------------------------------------------------------------

def create_submission_from_scores():
    """
    점수 리스트가 담긴 JSON을 읽어, 4개씩 그룹화한 후
    각 그룹에서 가장 큰 값의 위치를 A,B,C,D로 변환하여 CSV를 생성합니다.
    """
    
    # --- 1. JSON 파일 읽고 점수만 추출하기 ---
    try:
        with open(INPUT_JSON_FILE, 'r', encoding='utf-8') as f:
            raw_data = json.load(f)
        
        # JSON의 복잡한 구조 `[[[값]]]`에서 숫자 값만 추출하여 하나의 리스트로 만듭니다.
        scores = [item[0][0] for item in raw_data]
        print(f"성공적으로 '{INPUT_JSON_FILE}' 파일을 읽었습니다. 총 {len(scores)}개의 점수를 처리합니다.")
        
    except FileNotFoundError:
        print(f"오류: '{INPUT_JSON_FILE}' 파일을 찾을 수 없습니다. 파일 이름을 확인해주세요.")
        return
    except (TypeError, IndexError):
        print(f"오류: JSON 파일의 데이터 구조가 '[[[값]]]' 형식이 아닙니다. 파일 내용을 확인해주세요.")
        return
    except Exception as e:
        print(f"파일을 읽는 중 오류가 발생했습니다: {e}")
        return

    # --- 2. 4개씩 묶어서 정답(A,B,C,D) 결정 ---
    csv_rows = []
    question_index = 0
    choices = ['A', 'B', 'C', 'D']
    
    # range의 세 번째 인자(step)를 4로 주어 4칸씩 건너뛰며 반복
    for i in range(0, len(scores), 4):
        # 현재 그룹(4개)의 점수를 추출
        score_group = scores[i : i+4]
        
        # 그룹이 4개 미만이면 처리할 수 없으므로 중단
        if len(score_group) < 4:
            print(f"경고: 마지막 데이터 그룹이 4개 미만이므로 처리하지 않습니다. (남은 데이터 수: {len(score_group)})")
            break
            
        # ID 생성 (예: 0 -> TEST_000, 1 -> TEST_001)
        test_id = f"TEST_{question_index:03d}"
        
        # 그룹 내에서 가장 큰 값의 인덱스(위치)를 찾음 (0, 1, 2, 3 중 하나)
        # 예: max(score_group)이 그룹의 두 번째 값이라면 max_index는 1
        max_index = score_group.index(max(score_group))
        
        # 인덱스를 A, B, C, D로 변환
        answer = choices[max_index]
        
        # CSV의 한 줄을 리스트로 만들어 추가
        csv_rows.append([test_id, answer])
        
        # 다음 문제 번호를 위해 1 증가
        question_index += 1

    # --- 3. CSV 파일로 저장 ---
    try:
        with open(OUTPUT_CSV_FILE, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['ID', 'answer'])
            writer.writerows(csv_rows)
            
        print(f"성공적으로 '{OUTPUT_CSV_FILE}' 파일을 생성했습니다.")

    except Exception as e:
        print(f"CSV 파일을 저장하는 중 오류가 발생했습니다: {e}")


# 스크립트 실행
if __name__ == '__main__':
    create_submission_from_scores()
```
